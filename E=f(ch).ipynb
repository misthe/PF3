```python
import numpy as np
from matplotlib import pyplot as plt
import scienceplots
from scipy.optimize import curve_fit
```

```python
plt.style.use(['science','notebook','grid'])
```

```python
ch=[33,349]
E=[32,662]
```

```python
def model_f(x,a,b):
    return a*x+b
```

```python
popt,pcov=curve_fit(model_f,ch,E,p0=[2,-20])
```

```python
popt
```

```python
pcov
```

```python
a_opt,b_opt=popt
x=np.linspace(min(ch),max(ch),100)
y=model_f(x,a_opt,b_opt)
```

```python
textstr1='\n'.join((
    r'$E=\alpha \cdot ch + \beta$',
    r'$\alpha=%.2f$keV' % (popt[0]),
    r'$\beta=%.2f$keV' % (popt[1])
))
```

```python
plt.scatter(ch,E,color='black')
plt.plot(x,y,color='purple')
plt.title("Calibration")
plt.xlabel('Channels')
plt.ylabel('Energy[keV]')
plt.text(30,600,textstr1,bbox=dict(facecolor='white',edgecolor='black'))
```
